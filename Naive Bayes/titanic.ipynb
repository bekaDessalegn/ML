```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
df = pd .read_csv("titanic.csv")
df.head()
```

```python
df.drop(["PassengerId", "Name", "SibSp", "Parch", "Ticket", "Cabin", "Embarked"], axis="columns")
```

```python
target = df.Survived
target.head()
```

```python
inputs = df.drop(["Survived", "PassengerId", "Name", "SibSp", "Parch", "Ticket", "Cabin", "Embarked"], axis="columns")
inputs.head()
```

```python
dummy = pd.get_dummies(inputs[["Sex"]])
dummy.head()
```

```python
inputs = pd.concat([inputs, dummy], axis="columns")
inputs.head()
```

```python
inputs = inputs.drop(["Sex"], axis="columns")
inputs.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```
